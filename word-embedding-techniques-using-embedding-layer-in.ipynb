# Word Embedding Techniques using Embedding Layer in Keras

```python
import tensorflow
import keras
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = ['the glass of milk', 'The glass of juice', 'the cup of tea', 'I am a good boy', 
       'I am a good developer', 'understand the meaning of the words', 'your videos are good',
        'Hi my name is sushant twayana and i am studying engineering'
       ]
```

```python
sent
```

```python
# Vocabulary size
voc_size = 10000
```

**One Hot Representation -> Provides the index size for each and every word in the sentence**

```python
onehot_repr = [one_hot(words, voc_size)for words in sent]
print(onehot_repr)
```

**Onehot Repre ----> Embedding Layer in Keras**

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

Here, each word in the embedding must be represented by the same word lenght size. In this we consider the word_length_max = 8

```python
sent_length = 8 # maximum length of the word that is to be represented
embedded_docs = pad_sequences(onehot_repr,padding = 'pre', maxlen = sent_length)
print(embedded_docs)
```

```python
dim = 10 #features

model = Sequential()
model.add(Embedding(voc_size, 10, input_length = sent_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
# word converted into vector maps
print(model.predict(embedded_docs))
```
